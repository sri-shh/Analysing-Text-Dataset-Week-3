```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\srish\OneDrive\Documents\Internships\Millipixels\Week 3-4\Analysing-Text-Dataset-Week-3\ai_vs_human_content.csv")
```

```python
duplicate_count = df.duplicated().sum()
```

```python
print("Total duplicate rows:", duplicate_count)
```

```python
duplicate_content = df['content'].duplicated().sum()
```

```python
print("Duplicate content entries:", duplicate_content)
```

```python
empty_text = df[df['content'].str.strip() == '']
```

```python
print("Number of empty text entries:", len(empty_text))
```

```python
short_texts = df[df['word_count'] < 5]
```

```python
print("\nNumber of very short texts (less than 5 words):", len(short_texts))
```

```python
long_texts = df[df['word_count'] > 200]
```

```python
print("Number of very long texts (more than 200 words):", len(long_texts))
```
