```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from wordcloud import WordCloud
```

```python
#Word Frequencies
df = pd.read_csv(r"C:\Users\srish\OneDrive\Documents\Internships\Millipixels\Week 3-4\Analysing-Text-Dataset-Week-3\ai_vs_human_content.csv")
```

```python

all_text = ' '.join(df['content'].astype(str)).lower()
```

```python
all_words = pd.Series(all_text.split())
```

```python
word_freq = all_words.value_counts()
```

```python
print("Top 20 most common words:")
print(word_freq.head(20))
```

```python
print("\nWord Count Statistics:")
print(df['word_count'].describe())
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['word_count'], bins=30, color='steelblue', edgecolor='black')
plt.title("Text Length Distribution")
plt.xlabel("Word Count")
plt.ylabel("Number of Entries")
plt.show()
```

```python
#Word Cloud
all_content = ' '.join(df['content'].astype(str))
```

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_content)
```

```python
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud of All Content")
plt.show()
```
